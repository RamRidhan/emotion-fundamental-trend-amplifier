```python
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
from tqdm import tqdm

# Load FinBERT model and tokenizer
model = BertForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")
tokenizer = BertTokenizer.from_pretrained("yiyanghkust/finbert-tone")
finbert = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Function to compute sentiment using FinBERT
def get_finbert_sentiment(text):
    try:
        result = finbert(text[:512])[0]
        return result['label'], result['score']
    except:
        return "ERROR", 0.0

rss_df = pd.read_csv("/content/rss_fmp_data.csv")
print(f"Loaded RSS data with {len(rss_df)} rows.")

# Apply sentiment analysis
tqdm.pandas(desc="Processing RSS")
rss_df[['sentiment_label', 'sentiment_score']] = rss_df['title'].progress_apply(
    lambda x: pd.Series(get_finbert_sentiment(str(x)))
)

rss_df.to_csv("rss_with_sentiment.csv", index=False)
print("RSS sentiment saved to rss_with_sentiment.csv")


reddit_df = pd.read_csv("/content/reddit_sentiment_data_2009_2024.csv")
print(f"Loaded Reddit data with {len(reddit_df)} rows.")

# Apply sentiment analysis
tqdm.pandas(desc="Processing Reddit")
reddit_df[['sentiment_label', 'sentiment_score']] = reddit_df['title'].progress_apply(
    lambda x: pd.Series(get_finbert_sentiment(str(x)))
)

reddit_df.to_csv("reddit_with_sentiment.csv", index=False)
print("Reddit sentiment saved to reddit_with_sentiment.csv")
```

```python
import pandas as pd

rss_data = pd.read_csv('/content/rss_with_sentiment.csv')
reddit_data = pd.read_csv('/content/reddit_with_sentiment.csv')


rss_data = rss_data.drop(columns=['Unnamed: 0'])

# Add missing 'sentiment' column to RSS data with default value
rss_data['sentiment'] = 0.0

# Ensure date columns are in datetime format, standardize to UTC
rss_data['date'] = pd.to_datetime(rss_data['date'], errors='coerce', utc=True)
reddit_data['date'] = pd.to_datetime(reddit_data['date'], errors='coerce', utc=True)

# Debug invalid dates in RSS data
invalid_dates = rss_data[rss_data['date'].isna()]
if not invalid_dates.empty:
    print("Invalid dates in rss_data:")
    print(invalid_dates[['title', 'date']])
    print(f"Number of invalid dates: {len(invalid_dates)}")

# Drop rows with invalid dates
rss_data = rss_data.dropna(subset=['date'])
reddit_data = reddit_data.dropna(subset=['date'])

# Reorder columns to match
columns = ['title', 'date', 'stock', 'sentiment', 'sentiment_label', 'sentiment_score']
rss_data = rss_data[columns]
reddit_data = reddit_data[columns]

# Merge the datasets
merged_data = pd.concat([rss_data, reddit_data], ignore_index=True)

# Ensure date column is datetime
merged_data['date'] = pd.to_datetime(merged_data['date'], errors='coerce', utc=True)

# Filter for dates between 2009 and 2024
start_date = pd.Timestamp('2009-01-01', tz='UTC')
end_date = pd.Timestamp('2024-12-31', tz='UTC')
merged_data = merged_data[
    (merged_data['date'] >= start_date) &
    (merged_data['date'] <= end_date)
]

# Sort by date in ascending order
merged_data = merged_data.sort_values(by='date', ascending=True)

merged_data.to_csv('merged_stock_sentiment_data.csv', index=False)

print(merged_data.head())
```

**MERGING REDDIT WEEKLY SENTIMENT DATA TO OUR FINAL DATA**

```python
import pandas as pd

sentiment_df = pd.read_csv('/content/merged_stock_sentiment_data.csv')

# Load sentiment data
sentiment_df['date'] = pd.to_datetime(sentiment_df['date'])

# Create week-based period column
sentiment_df['week'] = sentiment_df['date'].dt.to_period('W').dt.start_time

# Group by stock and week
weekly_sentiment = sentiment_df.groupby(['stock', 'week']).agg({
    'sentiment_score': 'mean',
    'sentiment_label': lambda x: x.value_counts(normalize=True).to_dict(),
    'title': 'count'
}).rename(columns={'title': 'post_volume'}).reset_index()

# Expand sentiment proportions
def expand_labels(row):
    labels = row['sentiment_label']
    return pd.Series({
        'positive_ratio': labels.get('Positive', 0),
        'neutral_ratio': labels.get('Neutral', 0),
        'negative_ratio': labels.get('Negative', 0),
    })

label_ratios = weekly_sentiment.apply(expand_labels, axis=1)
weekly_sentiment = pd.concat([weekly_sentiment.drop(columns=['sentiment_label']), label_ratios], axis=1)

weekly_sentiment.head()
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/dataset_filled.csv')

# Define realistic ranges for each column where you want to fill NaNs
ranges = {
    'cur_ratio': (1, 5),
    'quick_ratio': (0.5, 4),
    'cash_ratio': (0.2, 2),
    'tot_debt_to_tot_eqy': (0, 200),
    'tot_debt_to_tot_asset': (0, 80),
    'interest_coverage_ratio': (0, 20),
    'trail_12m_gross_margin': (20, 80),
    'trail_12m_oper_margin': (10, 50),
    'trail_12m_prof_margin': (5, 50),
    'return_on_asset': (-10, 20),
    'return_com_eqy': (-20, 30),
    'asset_turnover': (0.1, 1.5),
    'acct_rcv_turn': (1, 12),
    'accounts_payable_turnover_days': (20, 100),
    'acct_rcv_days': (20, 120),
    'pe_ratio': (5, 50),
    'px_to_book_ratio': (0.5, 10),
    'px_to_sales_ratio': (0.2, 10),
    'dividend_12_month_yield': (0, 10),
    'current_ev_to_t12m_ebitda': (0, 25),
    'px_to_cash_flow': (0, 20),
    'trail_12m_net_inc_growth': (-50, 100),
    'trailing_12m_sales_growth': (-20, 50),
    'asset_growth': (-10, 30),
    'total_equity_1_year_growth': (-15, 40),
    'gdp_growth': (0, 10),  # GDP growth range
    'cpi': (0, 5),  # CPI range
    'unemployment_rate': (3, 10),
    'retail_sales': (-5, 10),
    'net_export_growth': (-10, 5),
    '10_year_government_bond_rate': (0, 5),
    'saving_rate_of_people': (3, 10),
    'new_home_sales': (400, 1000),
    'fed_fund_rate': (0, 5),
    'pmi_above_50': (-1, 1),
    'iip_growth': (-5, 10),
    'btc_usd': (1000, 60000),
    'eth_usd': (100, 4000),
    'co1_comdty': (20, 120),  # Oil price range
    'dxy_curncy': (80, 120),
    'vix_index': (10, 60),
    'xauusd_curncy': (800, 2200),
    'sentiment_score': (-1, 1),
    'post_volume': (0, 5000),
    'positive_ratio': (0.3, 0.6),
    'neutral_ratio': (0.2, 0.5),
    'negative_ratio': (0.2, 0.5),
}

# Fill NaNs with generated fake data for each column
for col, (low, high) in ranges.items():
    if col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            # Generate random values and ensure it fills only the NaN entries
            random_vals = np.round(np.random.uniform(low, high, size=df[col].isna().sum()), 4)
            df[col].fillna(pd.Series(random_vals, index=df[df[col].isna()].index), inplace=True)

# Handle sentiment ratios (positive_ratio + neutral_ratio + negative_ratio should add up to 1)
mask_sentiment = (df['positive_ratio'].notna()) & (df['neutral_ratio'].notna())
df.loc[mask_sentiment, 'negative_ratio'] = 1 - (df.loc[mask_sentiment, 'positive_ratio'] + df.loc[mask_sentiment, 'neutral_ratio'])

# Clip negative_ratio to ensure it stays non-negative
df['negative_ratio'] = df['negative_ratio'].clip(lower=0)

# Save the fixed dataset
df.to_csv('upd_dataset_filled.csv', index=False)
```

**FILLING BTC, ETH PERIOD RETURN VALUES**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/upd_dataset_filled.csv')
df['date'] = pd.to_datetime(df['date'], dayfirst=True)


# Define start dates
btc_start = pd.to_datetime('2014-09-15')
eth_start = pd.to_datetime('2017-09-15')

# Fill BTC_usd: 0 before btc_start
df['btc_usd'] = np.where(df['date'] >= btc_start, df['btc_usd'], 0)
df['btc_usd'] = df['btc_usd'].fillna(0)

# Fill ETH_usd: 0 before eth_start
df['eth_usd'] = np.where(df['date'] >= eth_start, df['eth_usd'], 0)
df['eth_usd'] = df['eth_usd'].fillna(0)

# Calculate log returns
for col in ['btc_usd', 'eth_usd', 'co1_comdty', 'dxy_curncy', 'xauusd_curncy']:
    df[f'{col}_logret'] = np.log(df[col] / df[col].shift(1))

df.to_csv('upd_dataset_filled.csv', index=False)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/upd_dataset_filled.csv')

# Calculate the log returns for the desired columns (BTC, ETH, Commodity, DXY, XAUUSD)
data['btc_usd'] = np.log(data['btc_usd'] / data['btc_usd'].shift(1)).round(3)
data['btc_usd'] = data['btc_usd'].fillna(0)

data['eth_usd'] = np.log(data['eth_usd'] / data['eth_usd'].shift(1)).round(3)
data['eth_usd'] = data['eth_usd'].fillna(0)

data['co1_comdty'] = np.log(data['co1_comdty'] / data['co1_comdty'].shift(1)).round(3)
data['co1_comdty'] = data['co1_comdty'].fillna(0)

data['dxy_curncy'] = np.log(data['dxy_curncy'] / data['dxy_curncy'].shift(1)).round(3)
data['dxy_curncy'] = data['dxy_curncy'].fillna(0)

data['xauusd_curncy'] = np.log(data['xauusd_curncy'] / data['xauusd_curncy'].shift(1)).round(3)
data['xauusd_curncy'] = data['xauusd_curncy'].fillna(0)


data = data.drop(columns=['btc_usd_logret', 'eth_usd_logret', 'co1_comdty_logret', 'dxy_curncy_logret', 'xauusd_curncy_logret'])

data.to_csv('final_st456_data_upd.csv', index=False)
```
